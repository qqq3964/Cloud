### SnapShot 생성
- snapshot은 내가 filestorage에 저장해놓은 과거로 돌아가도록 만들어주는 장치 그때의 시점으로 돌아가게 만들어주는것이다.

```python
# snapshot code
metadata={'foo':'bar'}

# 이코드를 실행하면 snapshot이 되는것을 확인 할 수 있다.
snapshot=file_service.snapshot_share('myshare',metadata=metadata)
```

```python
shares=list(file_service.list_shares(include_snapshots=True))
```

```python
for share in shares:
  print(share.snapshot)

# 원본은 None으로 나오고 그다음에 snapshot이 나온다.
```

```python
directories_and_files = list(
    file_service.list_directories_and_files(
        'myshare',
        snapshot='2023-01-26T02:33:31.0000000Z'
    )
)
```

```python
for file_or_dir in directories_and_files:
  print(file_or_dir.name)
```
